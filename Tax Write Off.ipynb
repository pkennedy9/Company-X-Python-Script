```python
# START WITH THE DATA AS SHOWN IN FIRST DATAFRAME. I GOT THIS INFO FROM OUR DATABASE
# AND THEN BY USING PANDAS, I CREATED ANOTHER DATAFRAME THAT I COULD UPLOAD BACK TO OUR DATABASE 

#------------------------------
#EXAMPLE OF DATA STARTED WITH
starting_data = pd.read_excel("/Users/peterkennedy/Desktop/Python scripts/companyX_work.xls")
starting_data
```

```python
pre_batch = pd.read_excel("/Users/peterkennedy/Desktop/Python scripts/companyX_work.xls")
n = 2
def key(test):    
    customer_num = test['Customer#']
    account_id = []
    for i in customer_num:
        account_id.extend([i] * n)
# account_id list corresponds to "Account Id" column in batch uploads

    invoice_num = test['Invoice #']
    inv_number = []
    for i in invoice_num:
        inv_number.extend([i] * n)
# inv_number list corresponds to"Inv Number" column in batch uploads 

    order_num = test['OrderNumber']
    order_number = []
    for i in order_num:
        order_number.extend([i] * n)
# order_number list corresponds to "Order Number" column in batch uploads

    x = test["AmountOutstanding"]
    postive_num = []
    negative_num = []
    for char in x:
        y = char.replace("(",'') 
        y = y.replace(")",'') 
        y = y.replace("$", '')
        y = "-" + y
        y = float(y)
        negative_num.append(y)
        #negative Tax Write Off amount
    for i in x:
        i = i.replace("(",'') 
        i = i.replace(")",'') 
        i = i.replace("$", '')
        i = float(i)
        postive_num.append(i)
        #postive state tax amount
    adj_amount = [i for j in zip(negative_num,postive_num) for i in j]
# adj_amount list corresponds to"Adj Amount" column in batch uploads

    state = test["TaxState"]
    adj_code = []
    for i in state:
        adj_code.append("TWO")
        adj_code.append("T" + i)
# adj_code list corresponds to "Adj Code" column in batch uploads 

# last two columns of batch uploads, "Adj Memo" and "Inv Note" are to be remained empty
    df = {"Account Id":account_id , "Inv Number": inv_number,  "Order Number": order_number,
          "Adj Amount": adj_amount,"Adj Code": adj_code, "Adj Memo":'', "Inv Note": ''}
    df = DataFrame(df)
    return df

# Function Ends --------------

df = key(pre_batch)
df.to_excel("/Users/peterkennedy/Desktop/Python Scripts/upload_file.xlsx", index = False)
```
